```python
from __future__ import division
from sympy import *
import sympy
```

```python
from sympy import diff, Symbol, sin, tan, integrate
x = Symbol('x')
i, n, m = symbols('i n m', integer=True)

integrate(x**10, (x, 0, n))
```

```python
from sympy import summation, oo, symbols, log
i, n, m = symbols('i n m', integer=True)
summation(x**10, (x, 0, n))
```

```python
summation((x-1)**10, (x, 0, n))
```

```python
integrate(((x-1)**10)*sympy.E**(-x), (x, 0, oo))
```

```python
y = Symbol('y')

f = symbols('f', cls = Function)
f = sympy.E**(-((x-1)**2+(y+1)**2))
f
```

```python
diff(f,x)
```

```python
diff(f,y)
```

```python
diff(diff(f,y),x)
```

```python
solve((diff(f,x), diff(f,y)), (x, y))
```

```python
from sympy.plotting import plot
p7=plotting.plot3d(f, show=False)
p7.show()
```
